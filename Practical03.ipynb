NumPy:

1.   NumPy is a powerful Python library used for numerical computations.
2.   It supports large, multi-dimensional arrays and matrices along with a collection of mathematical functions.
4.   It is ideal for simulations as it provides efficient random number generation for different probability distributions.

Matplotlib:
1.   Matplotlib is a plotting library used for creating static, animated, and interactive visualizations in Python.
2.   'pyplot' is a module in Matplotlib used to create figures and charts like line plots, bar charts, and histograms.


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Binomial Distribution (numpy.random.binomial):
# Parameters:
#     n = number of trials
#     p = probability of success in each trial
# Real-world Use-Case:
#     Tossing a coin 10 times and counting how many heads appear.

binomial_data = np.random.binomial(n=10, p=0.5, size=1000)
# Set up the plotting area with 3 subplots
plt.figure(figsize=(18, 5))

# Histogram for Binomial Distribution
plt.subplot(1, 3, 1)
plt.hist(binomial_data, bins=11, color='skyblue', edgecolor='black')
plt.title('Binomial Distribution (n=10, p=0.5)')
plt.xlabel('Number of Successes')
plt.ylabel('Frequency')
```

```python
# Normal Distribution (numpy.random.normal):
# Parameters:
#     mean (μ) = center of the distribution
#     standard deviation (σ) = spread of the distribution
# Features:
#     Symmetrical bell-shaped curve.
#     Commonly used to model natural phenomena (e.g., height, test scores).

normal_data = np.random.normal(loc=0, scale=1, size=1000)
# Set up the plotting area with 3 subplots
plt.figure(figsize=(18, 5))

# Histogram for Normal Distribution
plt.subplot(1, 3, 2)
plt.hist(normal_data, bins=30, color='lightgreen', edgecolor='black')
plt.title('Normal Distribution (μ=0, σ=1)')
plt.xlabel('Value')
plt.ylabel('Frequency')
```

```python
# Uniform Distribution (numpy.random.uniform):
# Parameters:
#     low = lower bound
#     high = upper bound
# Features:
#     All values in the interval [low, high] are equally likely to occur.

uniform_data = np.random.uniform(low=0, high=1, size=1000)
# Set up the plotting area with 3 subplots
plt.figure(figsize=(18, 5))

# Histogram for Uniform Distribution
plt.subplot(1, 3, 3)
plt.hist(uniform_data, bins=30, color='salmon', edgecolor='black')
plt.title('Uniform Distribution (0 to 1)')
plt.xlabel('Value')
plt.ylabel('Frequency')
```
